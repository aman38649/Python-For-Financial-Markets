```python
import talib
import yfinance as yf
```

```python
data = yf.download("TCS", start="2020-01-01", end="2020-11-01")
```

```python
data.head()
```

```python
morning_star = talib.CDLMORNINGSTAR(data['Open'], data['High'], data['Low'], data['Close'])
engulfing = talib.CDLENGULFING(data['Open'], data['High'], data['Low'], data['Close'])

data['Morning Star'] = morning_star
data['Engulfing'] = engulfing
```

```python
engulfing_days = data[data['Engulfing'] != 0]
engulfing_days
```

###In the above table -ve values shows **bearish** engulfing and +ve values shows **bullish** englfing pattern.
