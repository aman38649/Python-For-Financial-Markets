```python
import pandas_ta as ta
import pandas as pd
```

##1. T.A. On NSE - TCS

```python
ticker = yf.Ticker("TCS")
df = ticker.history(period="1y")
print(df)
```

```python
adx = df.ta.adx()
```

```python
print(adx)
```

```python
macd = df.ta.macd(fast=14, slow=28)

rsi = df.ta.rsi()

df = pd.concat([df, adx, macd, rsi], axis=1)

df
```

```python
temp_df = df[df['RSI_14'] < 30]

temp_df
```

##2. T.A. On Crypto Excahnge - Binance

```python
import ccxt
exchange = ccxt.binance()
```

```python
bars = exchange.fetch_ohlcv('ETH/USDT', timeframe='5m', limit=500)

df_crypto = pd.DataFrame(bars, columns=['time', 'open', 'high', 'low', 'close', 'volume'])

print(df_crypto)
```

```python
adx = df_crypto.ta.adx()
macd = df_crypto.ta.macd(fast=14, slow=28)
rsi = df_crypto.ta.rsi()

df_crypto = pd.concat([df_crypto, adx, macd, rsi], axis=1)
df_crypto
```

```python
fi = df_crypto[df_crypto['RSI_14'] < 30]
fi
```
