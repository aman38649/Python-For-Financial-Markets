## Get Last 1 Month Every 5 Minutes (Get 1 Minute Data for 7 Days)

```python
# Valid periods: 1d,5d,1mo,3mo,6mo,1y,2y,5y,10y,ytd,max
# Valid intervals: 1m,2m,5m,15m,30m,60m,90m,1h,1d,5d,1wk,1mo,3mo
tcs = yf.download(tickers='TCS.NS', period='1mo', interval='5m')
tcs
```

## Plot a Candlestick Chart

```python
x = tcs.index
close = tcs['Adj Close']
high = tcs['High']
low = tcs['Low']
openp = tcs['Open']

# Calculates 5 and 20 day moving average
# Takes prices and divides them by a defined time period
# Short periods react quickly, while longer react slower to price changes
# They are used to see if a stock is in an uptrend
# If a short term trend is above the longterm that is a sign an uptrend is expected
tcs['MA5'] = tcs.Close.rolling(5).mean()
tcs['MA20'] = tcs.Close.rolling(20).mean()

# Create candlestick plot
candles = go.Candlestick(x=x, open=openp, high=high,
                         low=low, close=close, name="Candles")

# Create 5 and 20 day moving average
ma5 = go.Scatter(x=tcs.index, y=tcs.MA5, 
                 line=dict(color='orange', width=1), name="MA5")
ma20 = go.Scatter(x=tcs.index, y=tcs.MA20, 
                  line=dict(color='green', width=1), name="MA20")

# Create volume bar chart
vol = go.Bar(x=tcs.index, y=tcs['Volume'], name="Volume")

# Create figure with secondary y-axis
fig = make_subplots(specs=[[{"secondary_y": True}]])

# Add plots
fig.add_trace(trace=candles, secondary_y=True)
fig.add_trace(trace=ma5, secondary_y=True)
fig.add_trace(trace=ma20, secondary_y=True)
fig.add_trace(trace=vol, secondary_y=False)

# Valid Steps : 'month', 'year', 'day', 'hour', 'minute', 'second', 'all'
fig.update_layout(
    xaxis=dict(
        rangeselector=dict(
            buttons=list([
                dict(count=15,
                     label="15M",
                     step="minute",
                     stepmode="backward"),
                dict(count=45,
                     label="45M",
                     step="minute",
                     stepmode="backward"),
                dict(count=1,
                     label="1HR",
                     step="hour",
                     stepmode="todate"),
                dict(count=1,
                     label="1D",
                     step="day",
                     stepmode="todate"),
                dict(count=3,
                     label="3D",
                     step="day",
                     stepmode="todate"),
                dict(count=7,
                     label="1W",
                     step="day",
                     stepmode="backward"),
                dict(count=14,
                     label="2W",
                     step="day",
                     stepmode="backward"),
                dict(count=1,
                     label="1M",
                     step="month",
                     stepmode="backward"),
                dict(label="All", step="all")
            ])
        ),
        rangeslider=dict(
            visible=True
        ),
        type="date"
    )
)
fig.update_layout(xaxis_title="Dates", yaxis_title="Stock Price", 
                  title="TCS Candlestick Chart", 
                  width=1000, height=800)

# Hides plot between saturday and monday
# Hide hours between 9:30am to 4pm
# Hide specific days like holidays
fig.update_xaxes(
        rangeslider_visible=True,
        rangebreaks=[
            dict(bounds=["sat", "mon"]),
            dict(bounds=[16, 9.5], pattern="hour"), 
            dict(values=["2020-12-25", "2021-01-01", "2021-07-04"])
        ]
    )

fig.show(renderer="colab")
```

## Simple Moving Average

A SMA allows you to see the big picture when analyzing a stock. It however it takes time to catch up to current trends. This lag for a 100 day moving average would be 100/2 or 50 days. Long term moving averages can be extremely important as we'll see when we look at the Death Cross and Golden Cross. 

```python
# Download tcs, BHARTIARTL.NS and the S&P
tcs_df = yf.download(tickers='TCS.NS', period='10y', interval='1d')
tcs_df
```

```python
artl_df = yf.download(tickers='BHARTIARTL.NS', period='10y', interval='1d')
artl_df
```

```python
nifty_df = yf.download(tickers='^NSEI', period='10y', interval='1d')
nifty_df
```

## Calculate 20 Day Moving Averages

```python
tcs_ma20 = tcs_df['Adj Close'].rolling(window=20).mean()
tcs_ma20.tail()
```

```python
artl_ma20 = artl_df['Adj Close'].rolling(window=20).mean()
artl_ma20.tail()
```

```python
nifty_ma20 = nifty_df['Adj Close'].rolling(window=20).mean()
nifty_ma20.tail()
```

## Calculate 100 Day Moving Average

```python
# Allows us to study the averaged out long term trends
tcs_ma100 = tcs_df['Adj Close'].rolling(window=100).mean()
artl_ma100 = artl_df['Adj Close'].rolling(window=100).mean()
nifty_ma100 = nifty_df['Adj Close'].rolling(window=100).mean()
nifty_ma100.tail()
```

## Plot tcs with Moving Averages

```python
ma20 = go.Scatter(x=tcs_ma20.index, y=tcs_ma20, 
                  line=dict(color='orange', width=1), name="MA20")
ma100 = go.Scatter(x=tcs_ma100.index, y=tcs_ma100, 
                   line=dict(color='green', width=1), name="MA100")
tcs_prc = go.Scatter(x=tcs_df.index, y=tcs_df['Adj Close'], 
                      line=dict(color='blue', width=1), name="Price")

fig = go.Figure()
fig.add_trace(ma20)
fig.add_trace(ma100)
fig.add_trace(tcs_prc)
                                              
fig.update_xaxes(
    rangeslider_visible=True, title='Zoom on Dates Using Slider')
fig.update_yaxes(title="Stock Price (USD)")
fig.show(renderer="colab")
```

## Exponential Moving Average

A EMA can be used to reduce the lag by putting more emphasis on recent price data. 

```python
# Calculate tcss 20 day EMA
tcs_ema20 = tcs_df['Adj Close'].ewm(span=20, adjust=False).mean()

ema20 = go.Scatter(x=tcs_ema20.index, y=tcs_ema20, 
                   line=dict(color='green', width=1), name="EMA20")
ma20 = go.Scatter(x=tcs_ma20.index, y=tcs_ma20, 
                  line=dict(color='orange', width=1), name="MA20")
tcs_prc = go.Scatter(x=tcs_df.index, y=tcs_df['Adj Close'], 
                      line=dict(color='blue', width=1), name="Price")

fig = go.Figure()
fig.add_trace(ma20)
fig.add_trace(ema20)
fig.add_trace(tcs_prc)
                                              
fig.update_xaxes(
    rangeslider_visible=True, title='Zoom on Dates Using Slider')
fig.update_yaxes(title="Stock Price (USD)")
fig.show(renderer="colab")

# As you can see a trading stategy where we buy when the price 
# hits the moving average from below signals a buy and vice versa.
```

## Death & Golden Crosses

When a Death Cross occurs, that is a sign that a major sell off will occur. A Death Cross is said to occur typically when the 50 day moving average falls below a 200 day. A Golden Cross accures when the short term average crosses the long term again moving higher.

```python
nifty_df = yf.download(tickers='^NSEI', period='max', interval='1d')
nifty_df
```

```python
nifty_ma50 = nifty_df['Adj Close'].rolling(window=50).mean()
nifty_ma200 = nifty_df['Adj Close'].rolling(window=200).mean()
```

## Plot NIFTY50

```python
ma50 = go.Scatter(x=nifty_ma50.index, y=nifty_ma50, 
                  line=dict(color='orange', width=1), name="MA50")
ma200 = go.Scatter(x=nifty_ma200.index, y=nifty_ma200, 
                   line=dict(color='green', width=1), name="MA200")
nifty_prc = go.Scatter(x=nifty_df.index, y=nifty_df['Adj Close'], 
                      line=dict(color='blue', width=1), name="Price")

fig = go.Figure()
fig.add_trace(ma50)
fig.add_trace(ma200)
fig.add_trace(nifty_prc)
                                              
fig.update_xaxes(
    rangeslider_visible=True, title='Zoom on Dates Using Slider')
fig.update_yaxes(title="Stock Price (USD)")
fig.show(renderer="colab")

# You can see many examples of death & golden crosses below
```
